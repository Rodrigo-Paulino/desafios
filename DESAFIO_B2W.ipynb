```python
# Importando bibliotecas, abrindo o arquivo, renomeando colunas:

import numpy as np
import pandas as pd

nomes_colunas = ['ID', 'Brand', 'Categories', 'Primary_Categories', 'Colors', 'Prices_Amount_Max', 'Prices_Amount_Min', 'Prices_Color', 'Prices_Condition', 'Prices_is_Sale']

sapatos = 'C:\\sapatos.csv'
sapatos_dados = pd.read_csv(sapatos, header = 0, names = nomes_colunas)
sapatos_dados.tail()
```

```python
# Conferindo o tipo dos dados:

sapatos_dados.dtypes
```

```python
# Limpeza de dados: Categories:

categories_split = []
new_categories = []

for i, cat in enumerate(sapatos_dados['Categories']):
    categories_split.append(cat.split(','))
    new_categories.append(categories_split[i][-1])

sapatos_dados['Categories'] = new_categories
```

```python
# Limpeza de dados: Primary_Categories

primary_categories_split = []
new_primary_categories = []

for i, cat in enumerate(sapatos_dados['Primary_Categories']):
    primary_categories_split.append(cat.split(','))
    new_primary_categories.append(primary_categories_split[i][-1])

sapatos_dados['Primary_Categories'] = new_primary_categories
```

```python
# Limpeza de dados: Colors

colors_split = []
new_colors = []

for i, col in enumerate(sapatos_dados['Colors']):
    if type(col) == float:
        colors_split.append('no color')
        new_colors.append(colors_split[i])
    else:
        colors_split.append(col.split(','))
        new_colors.append(colors_split[i][-1])

sapatos_dados['Colors'] = new_colors
```

```python
# Limpeza de dados: Prices_Amount_Max

# 1 - Substituir a ',' por '.'

sapatos_dados['Prices_Amount_Max'] = sapatos_dados.Prices_Amount_Max.str.replace(',','.')

# 2 - Transformar os dados em float.

Prices_Amount_Max_float = []      

for i in sapatos_dados.Prices_Amount_Max:
    Prices_Amount_Max_float.append(float(i))

sapatos_dados['Prices_Amount_Max'] = Prices_Amount_Max_float
```

```python
# Limpeza de dados: Condition:

condition_split = []
new_condition = []

for i, con in enumerate(sapatos_dados['Prices_Condition']):
    if type(con) == float:
        new_condition.append('no condition')
    else:
        new_condition.append(con.lower())
        
sapatos_dados['Prices_Condition'] = new_condition
```

```python
# Conferindo o novo tipo dos dados da coluna 'Prices_Amount_Max':

sapatos_dados.dtypes
```

```python
# Criando um agrupamento de dados nas colunas 'ID' e 'Prices_Color'.
# Calculando a 'Média' e o 'Desvio Padrão' das colunas 'Prices_Amount_Max' e 'Prices_Amount_Min'.


media_dpadrao_max = sapatos_dados.groupby(['ID','Prices_Color']).Prices_Amount_Max.agg(['mean','std'])
media_dpadrao_min = sapatos_dados.groupby(['ID','Prices_Color']).Prices_Amount_Min.agg(['mean','std'])

print('( Group = ID & Prices_Color ) + ( Amount Max = Média e Desvio Padão )')
print(media_dpadrao_max)

print('')
print('=' * 50)
print('')

print('( Group = ID & Prices_Color ) + ( Amount Mim - Média e Desvio Padão )')
print(media_dpadrao_min)
```

```python
# Criando um agrupamento de dados nas colunas 'ID' e 'Prices_Condition'.
# Calculando a 'Média' e o 'Desvio Padrão' das colunas 'Prices_Amount_Max' e 'Prices_Amount_Min'.

media_dpadrao_max = sapatos_dados.groupby(['ID','Prices_Condition']).Prices_Amount_Max.agg(['mean','std'])
media_dpadrao_min = sapatos_dados.groupby(['ID','Prices_Condition']).Prices_Amount_Min.agg(['mean','std'])

print('( Group = ID & Prices_Condition ) + ( Amount Max = Média e Desvio Padão )')
print(media_dpadrao_max)

print('')
print('-' * 45)
print('')

print('( Group = ID & Prices_Condition ) + ( Amount Mim - Média e Desvio Padão )')
print(media_dpadrao_min)
```

```python
# Criando um agrupamento de dados nas colunas 'ID' e 'Categories'.
# Calculando a 'Média' e o 'Desvio Padrão' das colunas 'Prices_Amount_Max' e 'Prices_Amount_Min'.

media_dpadrao_max = sapatos_dados.groupby(['ID','Categories']).Prices_Amount_Max.agg(['mean','std'])
media_dpadrao_min = sapatos_dados.groupby(['ID','Categories']).Prices_Amount_Min.agg(['mean','std'])

print('( Group = ID & Categories ) + ( Amount Max = Média e Desvio Padão )')
print(media_dpadrao_max)

print('')
print('-' * 45)
print('')

print('( Group = ID & Categories ) + ( Amount Mim - Média e Desvio Padão )')
print(media_dpadrao_min)
```
